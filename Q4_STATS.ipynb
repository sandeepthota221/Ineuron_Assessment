Correlation in Statistics

Correlation refers to the statistical association between two variables. It measures the direction and strength of the linear relationship between them. A correlation coefficient is calculated to quantify this relationship.

There are three main types of correlation:

Positive Correlation: When one variable increases, the other tends to increase as well. (e.g., Height and weight)

Negative Correlation: When one variable increases, the other tends to decrease. (e.g., Study time and test anxiety)

No Correlation: There's no clear linear relationship between the variables. (e.g., Shoe size and favorite color)


```python

# Sample data (replace with your actual data)
study_hours = [4, 6, 8, 5, 7, 3, 9, 2, 1, 10]
exam_scores = [65, 80, 92, 78, 85, 50, 95, 42, 38, 100]
```

```python
#Jupyter Notebook Code for Visualization:

# Python
import matplotlib.pyplot as plt
import numpy as np
# Calculate correlation coefficient (optional)
correlation = np.corrcoef(study_hours, exam_scores)[0, 1]  # Using NumPy

# Create scatter plot
plt.scatter(study_hours, exam_scores)

# Add labels and title
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study Hours vs Exam Scores")

# Add correlation coefficient (optional)
plt.text(0.7, 0.8, f"Correlation: {correlation:.2f}", ha='center', va='center')

# Display plot
plt.show()
```

Explanation of the Code:

Import libraries: We import matplotlib.pyplot for plotting and optionally numpy for calculating the correlation coefficient.
Sample Data: Replace the sample data with your actual data for study hours and exam scores.

Correlation Coefficient (Optional): This line calculates the correlation coefficient using NumPy's corrcoef function. You can uncomment this line to display the correlation value on the plot.

Scatter Plot: The plt.scatter function creates the scatter plot with study hours on the x-axis and exam scores on the y-axis.
Labels and Title: We add labels for the axes and a title for the plot.

Correlation Text (Optional): This line adds the calculated correlation coefficient to the plot for reference.
Display Plot: The plt.show() function displays the generated scatter plot.
